```python
import numpy as np

# Parameters for the normal distribution
mean_height = 175  # Mean height in centimeters
std_deviation = 10  # Standard deviation in centimeters

# Generating 20 random heights
random_heights = np.random.normal(mean_height, std_deviation, 10000)
```

```python
from scipy.stats import norm

# Generate points on the x axis (heights)
heights = np.linspace(mean_height - 3*std_deviation, mean_height + 3*std_deviation, 100)

# Calculate the PDF (Probability Density Function) of the normal distribution
pdf = norm.pdf(heights, mean_height, std_deviation)

# Creating the plot for the Gaussian curve
plt.figure(figsize=(10, 6))
plt.plot(heights, pdf, color='blue')

# Adding titles and labels
plt.title('Gaussian Curve for Normal Distribution of Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')

# Displaying the plot
plt.show()
```

```python
import numpy as np

# Parameters for the normal distribution
mean_height_2 = 164  # Mean height in centimeters
std_deviation_2 = 10  # Standard deviation in centimeters

# Generating 20 random heights
random_heights_2 = np.random.normal(mean_height, std_deviation, 10000)
```

```python
# Generate points on the x axis (heights)
heights = np.linspace(mean_height - 3*std_deviation, mean_height + 3*std_deviation, 100)
heights_2 = np.linspace(random_heights_2 - 3*std_deviation_2, random_heights_2 + 3*std_deviation_2, 100)

# Calculate the PDF for the second normal distribution
pdf = norm.pdf(heights, mean_height, std_deviation)
pdf_2 = norm.pdf(heights_2, mean_height_2, std_deviation_2)

# Creating the plot for both Gaussian curves
plt.figure(figsize=(10, 6))

# First Gaussian curve (original parameters)
plt.plot(heights, pdf, color='blue', label=f'Mean: {mean_height} cm, SD: {std_deviation} cm')

# Second Gaussian curve (new parameters)
plt.plot(heights_2, pdf_2, color='red', label=f'Mean: {mean_height_2} cm, SD: {std_deviation_2} cm')

# Adding titles, labels, and legend
plt.title('Comparison of Two Gaussian Curves for Different Height Distributions')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
# plt.legend()
```

```python
# Creating the plot
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))

# Plotting the distribution
sns.histplot(random_heights, kde=True)
# sns.histplot(random_heights_2, kde=True)

# Adding titles and labels
plt.title('Normal Distribution of Generated Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')

# Displaying the plot
plt.show()
```

```python
from scipy.stats import ttest_ind

# Sample data from the first distribution
sample_1 = random_heights

# Sample data from the second distribution
sample_2 = random_heights_2

# Performing the t-test
t_test_result = ttest_ind(sample_1, sample_2)
```
